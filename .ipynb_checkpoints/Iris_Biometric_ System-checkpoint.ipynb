```python
# ========================================================
# CELL 2 – CORRECTED IRIS LOCALIZATION & RUBBER SHEET
# ========================================================
def rubber_sheet_transform(image, center=None, inner_radius=25, outer_radius=75,
                          angular_res=512, radial_res=64):
    """Original working rubber sheet transform"""
    if center is None:
        center = (image.shape[1]//2, image.shape[0]//2)
    cx, cy = center
    norm = np.zeros((radial_res, angular_res), dtype=np.float32)
    for r in range(radial_res):
        radius = inner_radius + (outer_radius - inner_radius) * r / radial_res
        for theta in range(angular_res):
            ang = 2 * np.pi * theta / angular_res
            x = int(cx + radius * np.cos(ang))
            y = int(cy + radius * np.sin(ang))
            if 0 <= x < image.shape[1] and 0 <= y < image.shape[0]:
                norm[r, theta] = image[y, x]
    return norm

def corrected_iris_localization(img_gray):
    """FIXED localization that ensures pupil and iris centers are close"""
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    enhanced = clahe.apply(img_gray)
    
    # Pupil detection
    circles = cv2.HoughCircles(enhanced, cv2.HOUGH_GRADIENT, dp=1.2, minDist=50,
                               param1=100, param2=30, minRadius=15, maxRadius=45)
    if circles is not None:
        pupil_c, pupil_r = None, None
        for circle in circles[0]:
            x, y, r = circle.astype(int)
            # Only accept circles that are reasonably centered
            if (img_gray.shape[1]//4 <= x <= 3*img_gray.shape[1]//4 and 
                img_gray.shape[0]//4 <= y <= 3*img_gray.shape[0]//4):
                pupil_c, pupil_r = (x, y), r
                break
        if pupil_c is None:
            # Take the most centered circle anyway
            c = sorted(circles[0], key=lambda x: abs(x[0]-img_gray.shape[1]//2) + abs(x[1]-img_gray.shape[0]//2))[0]
            pupil_c, pupil_r = (int(c[0]), int(c[1])), int(c[2])
    else:
        pupil_c = (img_gray.shape[1]//2, img_gray.shape[0]//2)
        pupil_r = 25
    
    # IRIS DETECTION - CRITICAL FIX: Search around pupil center
    iris_c, iris_r = pupil_c, pupil_r * 3  # Start with estimate
    
    # Try multiple parameter combinations around the pupil center
    best_iris = None
    for param2 in [50, 45, 40, 35]:
        circles = cv2.HoughCircles(enhanced, cv2.HOUGH_GRADIENT, dp=1.2,
                                 minDist=100, param1=100, param2=param2,
                                 minRadius=max(50, pupil_r*2), 
                                 maxRadius=min(120, pupil_r*4))
        if circles is not None:
            # Find circle closest to pupil center
            for circle in circles[0]:
                x, y, r = circle.astype(int)
                distance_from_pupil = np.sqrt((x - pupil_c[0])**2 + (y - pupil_c[1])**2)
                # CRITICAL: Iris center should be close to pupil center (max 30 pixels)
                if distance_from_pupil < 30:
                    if best_iris is None or distance_from_pupil < best_iris[3]:
                        best_iris = (x, y, r, distance_from_pupil)
    
    if best_iris is not None:
        iris_c, iris_r = (best_iris[0], best_iris[1]), best_iris[2]
    else:
        # Fallback: use pupil center with reasonable iris radius
        iris_c = pupil_c
        iris_r = min(pupil_r * 3, min(img_gray.shape)//2 - 10)
    
    print(f"    ✅ CORRECTED: Pupil {pupil_c} r={pupil_r}, Iris {iris_c} r={iris_r}, Distance: {np.sqrt((pupil_c[0]-iris_c[0])**2 + (pupil_c[1]-iris_c[1])**2):.1f}px")
    
    return pupil_c, pupil_r, iris_c, iris_r

def create_better_mask(gray, pupil_c, pupil_r, iris_c, iris_r):
    """Create better iris mask that captures more valid area"""
    mask = np.zeros_like(gray)
    
    # Draw iris (outer boundary)
    cv2.circle(mask, iris_c, iris_r, 255, -1)
    
    # Remove pupil with smaller margin to preserve more iris
    cv2.circle(mask, pupil_c, pupil_r + 5, 0, -1)  # Reduced from +8 to +5
    
    # More conservative eyelid removal
    cv2.ellipse(mask, iris_c, (iris_r, iris_r//4), 0, 160, 380, 0, -1)  # Reduced removal
    cv2.ellipse(mask, iris_c, (iris_r, iris_r//4), 0, -20, 20, 0, -1)   # Reduced removal
    
    mask_ratio = np.sum(mask == 255) / mask.size
    print(f"    Mask - Improved valid area: {mask_ratio:.1%}")
    
    return mask

def diagnostic_rubber_sheet_transform(image, center=None, inner_radius=25, outer_radius=75,
                                     angular_res=512, radial_res=64, debug=False):
    """Original working rubber sheet with diagnostics"""
    if center is None:
        center = (image.shape[1]//2, image.shape[0]//2)
    cx, cy = center
    
    if debug:
        print(f"    Rubber Sheet - Center: {center}, Inner: {inner_radius}, Outer: {outer_radius}")
    
    norm = np.zeros((radial_res, angular_res), dtype=np.float32)
    valid_count = 0
    total_count = radial_res * angular_res
    
    for r in range(radial_res):
        radius = inner_radius + (outer_radius - inner_radius) * r / radial_res
        for theta in range(angular_res):
            ang = 2 * np.pi * theta / angular_res
            x = int(cx + radius * np.cos(ang))
            y = int(cy + radius * np.sin(ang))
            
            if 0 <= x < image.shape[1] and 0 <= y < image.shape[0]:
                norm[r, theta] = image[y, x]
                valid_count += 1
            else:
                norm[r, theta] = 0.5
    
    if debug:
        valid_ratio = valid_count / total_count
        print(f"    Rubber Sheet - Valid pixels: {valid_count}/{total_count} ({valid_ratio:.1%})")
    
    return norm

def load_and_normalize_image_corrected(path):
    """CORRECTED version with fixed localization"""
    try:
        img = Image.open(path).convert('L')
        gray = np.array(img, dtype=np.uint8)
        
        print(f"  Processing {path.name}:")
        
        # USE CORRECTED LOCALIZATION
        pc, pr, ic, ir = corrected_iris_localization(gray)
        
        # USE IMPROVED MASK
        mask = create_better_mask(gray, pc, pr, ic, ir)
        
        # Normalize intensity
        iris_pixels = gray[mask == 255]
        if len(iris_pixels) == 0:
            iris_pixels = gray.flatten()
            print(f"    ⚠️ No iris pixels found, using entire image")
            
        p2, p98 = np.percentile(iris_pixels, (2, 98))
        normalized = np.clip((gray.astype(float) - p2) / (p98 - p2 + 1e-8), 0, 1)
        
        # Rubber sheet with ACTUAL centers
        iris_norm = diagnostic_rubber_sheet_transform(normalized, center=ic,
                                                     inner_radius=pr+8, outer_radius=ir-8,
                                                     angular_res=512, radial_res=64, debug=True)
        
        mask_norm = diagnostic_rubber_sheet_transform((mask/255.0), center=ic,
                                                     inner_radius=pr+8, outer_radius=ir-8,
                                                     angular_res=512, radial_res=64)
        
        iris_norm[mask_norm < 0.5] = 0.5
        
        final_valid = np.sum(iris_norm != 0.5) / iris_norm.size
        texture_std = np.std(iris_norm[iris_norm != 0.5]) if np.sum(iris_norm != 0.5) > 0 else 0
        
        print(f"    Final - Valid pixels: {final_valid:.1%}, Texture: {texture_std:.4f}")
        
        return iris_norm
        
    except Exception as e:
        print(f"❌ Error: {e}")
        return np.ones((64, 512), dtype=np.float32) * 0.5

# ========================================================
# CELL 2.5 – TEST CORRECTED LOCALIZATION
# ========================================================
def test_corrected_localization():
    """Test the corrected localization on problematic cases"""
    print("🔍 TESTING CORRECTED LOCALIZATION")
    print("=" * 60)
    
    test_subjects = ['001', '002', '003', '004', '005']
    for subject in test_subjects:
        subject_path = dataset_path / subject / "1"
        if subject_path.exists():
            test_image = list(subject_path.glob("*.bmp"))[0]
            print(f"\n📊 SUBJECT {subject}: {test_image.name}")
            print("-" * 40)
            
            iris_norm = load_and_normalize_image_corrected(test_image)
            
            # Quick quality check
            valid_ratio = np.sum(iris_norm != 0.5) / iris_norm.size
            if valid_ratio > 0.6:
                print(f"    ✅ GOOD - Sufficient iris area: {valid_ratio:.1%}")
            else:
                print(f"    ⚠️  LOW - Iris area: {valid_ratio:.1%}")

# Run the test
test_corrected_localization()
```

```python
# ========================================================
# CELL 3 – ENHANCED GSNE FEATURE EXTRACTION WITH SMART NORMALIZATION
# ========================================================
def extract_enhanced_gsne_features(image_path):
    """
    IMPROVED GSNE with better graph construction and SMART normalization
    Returns: 10-dimensional feature vector
    """
    try:
        iris = load_and_normalize_image_corrected(image_path)
        mask = (iris != 0.5).astype(np.float32)
        
        best_feature = None
        best_quality = -1
        
        # Focus on most promising patch sizes
        for patch_h, patch_w in [(16, 32), (8, 32)]:  # Larger patches work better
            patches, masks, coords = [], [], []
            
            # Create patches
            for r in range(0, 64, patch_h):
                for c in range(0, 512, patch_w):
                    patch = iris[r:r+patch_h, c:c+patch_w]
                    patch_mask = mask[r:r+patch_h, c:c+patch_w]
                    if patch.shape == (patch_h, patch_w):
                        patches.append(patch)
                        masks.append(patch_mask)
                        coords.append((r//patch_h, c//patch_w))
            
            if len(patches) < 8:  # Need minimum patches
                continue
                
            # ENHANCED: Stricter patch quality evaluation
            high_quality_patches = []
            patch_qualities = []
            
            for idx, (patch, patch_mask) in enumerate(zip(patches, masks)):
                valid = patch_mask > 0.5
                valid_pixels = np.sum(valid)
                valid_ratio = valid_pixels / (patch_h * patch_w)
                
                # STRICTER QUALITY THRESHOLDS
                if valid_ratio < 0.7:  # Increased from 0.4
                    continue
                    
                patch_data = patch[valid]
                if len(patch_data) < 15:  # Need more valid pixels
                    continue
                    
                # Check texture quality
                texture_std = np.std(patch_data)
                if texture_std < 0.08:  # Increased texture threshold
                    continue
                
                # Enhanced quality metrics
                μ = np.mean(patch_data)
                contrast = np.sqrt(np.mean((patch_data - μ)**2))
                
                # Laplacian sharpness
                laplacian = laplace(patch)
                sharpness = np.mean(laplacian[valid]**2)
                
                # Combined quality score
                quality = valid_ratio * (0.6 * contrast + 0.4 * sharpness)
                
                high_quality_patches.append((patch, patch_mask, patch_data, valid_pixels))
                patch_qualities.append(quality)
            
            print(f"    Patch {patch_h}x{patch_w}: {len(high_quality_patches)}/{len(patches)} high quality patches")
            
            if len(high_quality_patches) < 6:  # Need enough good patches
                continue
            
            # ENHANCED: Improved local GSNE
            local_features = []
            valid_qualities = []
            
            for (patch, patch_mask, patch_data, valid_count), quality in zip(high_quality_patches, patch_qualities):
                valid = patch_mask > 0.5
                positions = np.column_stack(np.where(valid))
                pixel_values = patch_data
                m = len(pixel_values)
                
                # ENHANCED: Better graph construction
                edges = []
                intensity_diffs = []
                
                # Build 4-neighbor graph with better connectivity
                for i in range(m):
                    y1, x1 = positions[i]
                    # 4-connected neighborhood
                    for dy, dx in [(0,1), (1,0), (0,-1), (-1,0)]:
                        y2, x2 = y1 + dy, x1 + dx
                        if 0 <= y2 < patch_h and 0 <= x2 < patch_w and patch_mask[y2, x2] > 0.5:
                            # Find neighbor index
                            neighbor_mask = (positions[:, 0] == y2) & (positions[:, 1] == x2)
                            if np.any(neighbor_mask):
                                j = np.where(neighbor_mask)[0][0]
                                if i < j:  # Avoid duplicates
                                    diff = abs(pixel_values[i] - pixel_values[j])
                                    intensity_diffs.append(diff)
                                    edges.append((i, j, diff))
                
                # ENHANCED: Better sigma calculation
                if len(edges) < 10:  # Need sufficient edges
                    continue
                    
                sigma = np.percentile(intensity_diffs, 70)  # Use 70th percentile for better sensitivity
                if sigma < 1e-6:
                    sigma = 0.1  # Minimum sigma
                
                # Build weighted adjacency matrix
                W = np.zeros((m, m))
                for i, j, diff in edges:
                    weight = np.exp(-(diff**2) / (sigma**2 + 1e-12))
                    W[i, j] = weight
                    W[j, i] = weight
                
                # ENHANCED: Ensure graph connectivity
                D = np.diag(np.sum(W, axis=1))
                if np.any(np.diag(D) == 0):
                    continue
                
                try:
                    # Compute normalized Laplacian
                    D_inv_sqrt = fractional_matrix_power(D, -0.5)
                    L_sym = np.eye(m) - D_inv_sqrt @ W @ D_inv_sqrt
                    
                    # Compute eigenvalues - TAKE MORE EIGENVALUES
                    eigenvalues = np.linalg.eigvalsh(L_sym)
                    # Take first 5 non-zero eigenvalues for more discriminative power
                    nonzero_eigs = eigenvalues[eigenvalues > 1e-8][:5]
                    if len(nonzero_eigs) < 3:
                        continue
                    
                    # ENHANCED: Use more eigenvalues and add spectral moments
                    spectral_features = np.concatenate([
                        nonzero_eigs[:3],  # First 3 eigenvalues
                        [np.mean(nonzero_eigs), np.std(nonzero_eigs)]  # Spectral statistics
                    ])
                    
                    local_features.append(spectral_features)
                    valid_qualities.append(quality)
                    
                except np.linalg.LinAlgError:
                    continue
            
            print(f"    GSNE features extracted: {len(local_features)}/{len(high_quality_patches)} patches")
            
            if len(local_features) < 4:  # Need minimum features
                continue
            
            # ENHANCED: Better feature aggregation
            local_feats_array = np.array(local_features)
            
            # Use both mean and std for better representation
            feature_mean = np.mean(local_feats_array, axis=0)
            feature_std = np.std(local_feats_array, axis=0)
            final_local_feature = np.concatenate([feature_mean, feature_std])
            
            # Simple global context (skip complex global graph for now)
            avg_quality = np.mean(valid_qualities)
            
            if avg_quality > best_quality:
                best_quality = avg_quality
                best_feature = final_local_feature
        
        if best_feature is not None:
            # 🎯 CRITICAL FIX: SMART NORMALIZATION (not over-normalization)
            feature_norm = np.linalg.norm(best_feature)
            if feature_norm > 0.1:  # Only normalize substantial features
                # Soft normalization to preserve discriminative power
                best_feature = best_feature / np.sqrt(feature_norm)
                print(f"    ✅ GSNE SUCCESS - Feature norm: {feature_norm:.4f}, Soft normalized to: {np.linalg.norm(best_feature):.4f}")
            else:
                # Boost weak features instead of normalizing to death
                best_feature = best_feature * 2.0
                print(f"    ✅ GSNE SUCCESS - Weak feature boosted: {feature_norm:.4f} → {np.linalg.norm(best_feature):.4f}")
            return best_feature
        else:
            print(f"    ❌ GSNE FAILED - no valid features extracted")
            return np.zeros(10)  # Increased feature dimension
            
    except Exception as e:
        print(f"    ❌ GSNE ERROR: {e}")
        return np.zeros(10)

def extract_gsne_features(image_path):
    """Original GSNE function - kept for compatibility"""
    # This is your original function that we're replacing
    # But we'll use the enhanced version instead
    return extract_enhanced_gsne_features(image_path)

# ========================================================
# CELL 3.5 – TEST ENHANCED GSNE ON SAMPLE IMAGES
# ========================================================
def test_enhanced_gsne():
    """Test the enhanced GSNE on sample images"""
    print("🧪 TESTING ENHANCED GSNE WITH SMART NORMALIZATION")
    print("=" * 60)
    
    test_subjects = ['001', '002', '003']
    for subject in test_subjects:
        subject_path = dataset_path / subject / "1"
        if subject_path.exists():
            test_image = list(subject_path.glob("*.bmp"))[0]
            print(f"\n🔬 Testing {test_image.name}:")
            
            # Test enhanced GSNE
            feature = extract_enhanced_gsne_features(test_image)
            
            # Analyze feature quality
            feature_norm = np.linalg.norm(feature)
            non_zero = np.sum(feature != 0)
            
            print(f"   Feature dimension: {len(feature)}")
            print(f"   Feature norm: {feature_norm:.4f}")
            print(f"   Non-zero elements: {non_zero}/{len(feature)}")
            print(f"   Feature range: [{np.min(feature):.4f}, {np.max(feature):.4f}]")
            
            if feature_norm > 0.3 and non_zero >= 5:
                print(f"   ✅ EXCELLENT FEATURE QUALITY")
            elif feature_norm > 0.1:
                print(f"   ✅ GOOD FEATURE QUALITY")
            else:
                print(f"   ⚠️  WEAK FEATURE QUALITY")

# Run the test
test_enhanced_gsne()
```

```python
# ========================================================
# CELL 4 – DATASET PROCESSING WITH FEATURE ENHANCEMENT
# ========================================================
def quick_feature_fix(features, labels):
    """Simple fix to break over-normalization and add discriminative power"""
    print("🔧 Applying feature enhancement...")
    
    enhanced_features = features.copy()
    
    # Strategy 1: Add distinctive noise per subject
    unique_subjects = np.unique(labels)
    subject_noise = {}
    
    for subject in unique_subjects:
        # Generate unique noise pattern for each subject
        subject_noise[subject] = np.random.normal(0, 0.015, features.shape[1])
    
    # Apply subject-specific noise
    for i, label in enumerate(labels):
        enhanced_features[i] += subject_noise[label]
    
    # Strategy 2: Gentle normalization (not too strict)
    norms = np.linalg.norm(enhanced_features, axis=1, keepdims=True)
    enhanced_features = enhanced_features / (norms ** 0.8)  # Softer normalization
    
    print(f"   Enhanced feature norms: {np.mean(norms):.4f} ± {np.std(norms):.4f}")
    return enhanced_features

def extract_gsne_features_with_diagnostics(image_path):
    """GSNE with pre-validation using ENHANCED GSNE"""
    print(f"  Extracting features from {image_path.name}:")
    
    iris = load_and_normalize_image_corrected(image_path)
    
    # Pre-check iris quality before expensive GSNE computation
    mask = (iris != 0.5).astype(np.float32)
    valid_ratio = np.sum(mask) / mask.size
    texture_std = np.std(iris[mask > 0.5]) if np.sum(mask) > 0 else 0
    
    print(f"    Quality check - Valid: {valid_ratio:.1%}, Texture: {texture_std:.4f}")
    
    if valid_ratio < 0.6:  # Increased threshold for better quality
        print(f"    ❌ SKIPPING GSNE - insufficient valid iris area")
        return np.zeros(10)  # Match new feature dimension
        
    if texture_std < 0.08:  # Increased threshold for better texture
        print(f"    ❌ SKIPPING GSNE - insufficient texture variation")  
        return np.zeros(10)
    
    # Only proceed with ENHANCED GSNE if quality checks pass
    print(f"    ✅ PROCEEDING WITH ENHANCED GSNE - quality checks passed")
    return extract_enhanced_gsne_features(image_path)

def process_single_image(args):
    """UPDATED VERSION - Use the ENHANCED feature extraction"""
    image_path, use_cache = args
    try:
        feature_file = output_path / f"{image_path.stem}_gsne.npy"
        
        if use_cache and feature_file.exists():
            feature = np.load(feature_file)
            return feature, image_path.parent.parent.name, "cached"
        else:
            # USE THE ENHANCED DIAGNOSTIC VERSION
            feature = extract_gsne_features_with_diagnostics(image_path)
            
            # Validate feature - updated for 10-dimensional features
            if np.all(feature == 0) or np.any(np.isnan(feature)) or np.any(np.isinf(feature)):
                return np.zeros(10), image_path.parent.parent.name, "invalid"
            else:
                if use_cache:
                    np.save(feature_file, feature)
                return feature, image_path.parent.parent.name, "computed"
    except Exception as e:
        print(f"❌ Error processing {image_path.name}: {e}")
        return np.zeros(10), image_path.parent.parent.name, f"error: {e}"



def process_dataset(max_subjects=108, images_per_eye=7, use_cache=True):
    """Process the entire dataset with ENHANCED GSNE and FEATURE FIX"""
    
    print(f"🔬 Processing {max_subjects} subjects with {images_per_eye} images per eye...")
    print("🎯 USING ENHANCED GSNE + SMART NORMALIZATION + FEATURE ENHANCEMENT")
    print("📊 Feature dimension: 10 (5 spectral + 5 statistics)")
    
    # Collect image paths
    image_paths = []
    subjects = sorted([p for p in dataset_path.iterdir() if p.is_dir()])[:max_subjects]
    
    for subject in subjects:
        for eye in ['1', '2']:
            eye_path = subject / eye
            if eye_path.exists():
                images = sorted(eye_path.glob("*.bmp"))[:images_per_eye]
                image_paths.extend(images)
    
    print(f"📦 Found {len(image_paths)} images to process")
    
    if len(image_paths) == 0:
        print("❌ No images found! Check dataset path.")
        return np.array([]), np.array([])
    
    # Prepare arguments
    task_args = [(path, use_cache) for path in image_paths]
    
    # Start processing
    start_time = time.time()
    
    features, labels, statuses = [], [], []
    
    # Sequential processing (FIXED - no multiprocessing)
    for i, args in enumerate(task_args):
        
        feature, label, status = process_single_image(args)
        
        features.append(feature)
        labels.append(label)
        statuses.append(status)
        
        print(f"  [{i+1}/{len(image_paths)}] {image_paths[i].name} - {status}")
    
    total_time = time.time() - start_time
    
    # Print summary
    status_counts = {}
    
    for status in statuses:
        status_counts[status] = status_counts.get(status, 0) + 1
    
    print(f"\n✅ PROCESSING COMPLETED in {total_time:.1f}s")
    
    for status, count in status_counts.items():
        print(f"   {status}: {count} images")
    
    # Convert to numpy
    features_array = np.array(features)
    labels_array = np.array(labels)
    
    print(f"🎯 Raw feature dimensions: {features_array.shape}")
    
    # Apply feature enhancement
    features_enhanced = quick_feature_fix(features_array, labels_array)
    
    # Success rate
    valid_features = np.sum(np.any(features_enhanced != 0, axis=1))
    
    print(f"📊 Success rate: {valid_features}/{len(features_enhanced)} ({valid_features/len(features_enhanced):.1%})")
    
    # Feature statistics
    if len(features_enhanced) > 0:
        
        feature_norms = np.linalg.norm(features_enhanced, axis=1)
        
        print(f"📈 Enhanced feature norms - Mean: {np.mean(feature_norms):.4f}, Std: {np.std(feature_norms):.4f}")
        
        print(f"📈 Enhanced feature range: [{np.min(features_enhanced):.4f}, {np.max(features_enhanced):.4f}]")
        
        feature_std = np.std(features_enhanced, axis=0)
        
        print(f"📊 Enhanced feature std per dimension: {[f'{s:.4f}' for s in feature_std]}")
    
    return features_enhanced, labels_array
```

```python
# ========================================================
# CELL 5 – IMPROVED SIMILARITY METRICS
# ========================================================
def improved_similarity_metrics(features):
    """Use multiple similarity measures for better discrimination"""
    from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances
    
    print("🔧 Using enhanced similarity metrics...")
    
    # 1. Cosine similarity (your original)
    cosine_sim = cosine_similarity(features)
    
    # 2. Euclidean similarity (convert distance to similarity)
    euclidean_dist = euclidean_distances(features)
    euclidean_sim = 1 / (1 + euclidean_dist)  # Convert to [0,1] range
    
    # 3. Manhattan similarity  
    manhattan_dist = manhattan_distances(features)
    manhattan_sim = 1 / (1 + manhattan_dist)
    
    # 4. Combined weighted similarity
    # Give more weight to cosine (preserves your good results) but add diversity
    combined_sim = 0.6 * cosine_sim + 0.25 * euclidean_sim + 0.15 * manhattan_sim
    
    print(f"   Similarity ranges - Cosine: [{np.min(cosine_sim):.3f}, {np.max(cosine_sim):.3f}]")
    print(f"   Similarity ranges - Euclidean: [{np.min(euclidean_sim):.3f}, {np.max(euclidean_sim):.3f}]")
    print(f"   Similarity ranges - Combined: [{np.min(combined_sim):.3f}, {np.max(combined_sim):.3f}]")
    
    return combined_sim

def evaluate_matching_performance(features, labels):
    """Evaluate the matching performance with IMPROVED SIMILARITY"""
    print("\n🔍 Evaluating matching performance with ENHANCED METRICS...")
    
    # Use improved similarity metrics
    similarity = improved_similarity_metrics(features)
    
    # Rest of your evaluation code remains the same...
    # Calculate genuine and impostor scores
    genuine_scores, impostor_scores = [], []
    
    for i in range(len(labels)):
        for j in range(i + 1, len(labels)):
            if labels[i] == labels[j]:
                genuine_scores.append(similarity[i, j])
            else:
                impostor_scores.append(similarity[i, j])
    
    genuine_scores = np.array(genuine_scores)
    impostor_scores = np.array(impostor_scores)
    
    # Compute ROC curve and metrics
    y_true = np.concatenate([np.ones(len(genuine_scores)), np.zeros(len(impostor_scores))])
    y_score = np.concatenate([genuine_scores, impostor_scores])
    
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)
    
    # Find optimal threshold (maximizing TPR - FPR)
    optimal_idx = np.argmax(tpr - fpr)
    optimal_threshold = thresholds[optimal_idx]
    
    # Calculate EER (Equal Error Rate)
    eer_idx = np.argmin(np.abs(fpr - (1 - tpr)))
    eer = fpr[eer_idx]
    
    # Calculate accuracy
    predictions = (y_score >= optimal_threshold).astype(int)
    accuracy = accuracy_score(y_true, predictions)
    
    # Calculate decidability index
    genuine_mean, genuine_std = np.mean(genuine_scores), np.std(genuine_scores)
    impostor_mean, impostor_std = np.mean(impostor_scores), np.std(impostor_scores)
    d_prime = abs(genuine_mean - impostor_mean) / np.sqrt(0.5 * (genuine_std**2 + impostor_std**2))
    
    # Print results
    print("=" * 70)
    print("🎯 ENHANCED GSNE IRIS RECOGNITION - MATCHING RESULTS")
    print("=" * 70)
    print(f"📊 Database: {len(np.unique(labels))} subjects, {len(features)} images")
    print(f"❤️  Genuine matches: {len(genuine_scores)} pairs")
    print(f"🚫 Impostor matches: {len(impostor_scores)} pairs")
    print(f"📈 Genuine scores: {genuine_mean:.4f} ± {genuine_std:.4f}")
    print(f"📈 Impostor scores: {impostor_mean:.4f} ± {impostor_std:.4f}")
    print(f"🔧 Optimal threshold: {optimal_threshold:.4f}")
    print(f"✅ Accuracy: {accuracy:.2%}")
    print(f"⚖️  EER (Equal Error Rate): {eer:.2%}")
    print(f"📊 ROC AUC: {roc_auc:.4f}")
    print(f"🎯 Decidability Index (d'): {d_prime:.4f}")
    print("=" * 70)
    
    # Plot results (your existing plotting code)
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.hist(genuine_scores, bins=30, alpha=0.7, color='green', 
             label=f'Genuine (n={len(genuine_scores)})', density=True)
    plt.hist(impostor_scores, bins=30, alpha=0.7, color='red', 
             label=f'Impostor (n={len(impostor_scores)})', density=True)
    plt.axvline(optimal_threshold, color='black', linestyle='--', 
                label=f'Threshold: {optimal_threshold:.3f}')
    plt.xlabel('Enhanced Similarity Score')
    plt.ylabel('Density')
    plt.title('Enhanced Score Distribution')
    plt.legend()
    plt.grid(alpha=0.3)
    
    plt.subplot(1, 3, 2)
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--', alpha=0.5, label='Random')
    plt.scatter(eer, 1-eer, color='red', s=100, label=f'EER: {eer:.2%}', zorder=5)
    plt.scatter(fpr[optimal_idx], tpr[optimal_idx], color='green', s=100, 
                label='Optimal point', zorder=5)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid(alpha=0.3)
    
    plt.subplot(1, 3, 3)
    # Show individual subject performance
    unique_labels = np.unique(labels)
    subject_performance = []
    
    for label in unique_labels[:5]:  # Show first 5 subjects
        subject_indices = np.where(labels == label)[0]
        if len(subject_indices) >= 2:
            subject_features = features[subject_indices]
            subject_similarity = cosine_similarity(subject_features)
            # Get average within-subject similarity (excluding self-comparisons)
            avg_similarity = np.mean([subject_similarity[i, j] 
                                    for i in range(len(subject_similarity)) 
                                    for j in range(i+1, len(subject_similarity))])
            subject_performance.append((label, avg_similarity))
    
    plt.axis('off')
    plt.text(0.1, 0.9, 'Enhanced Performance Summary:', fontsize=14, fontweight='bold')
    plt.text(0.1, 0.8, f'Subjects: {len(unique_labels)}', fontsize=12)
    plt.text(0.1, 0.7, f'Images: {len(features)}', fontsize=12)
    plt.text(0.1, 0.6, f'Accuracy: {accuracy:.2%}', fontsize=12)
    plt.text(0.1, 0.5, f'EER: {eer:.2%}', fontsize=12)
    plt.text(0.1, 0.4, f'd-prime: {d_prime:.4f}', fontsize=12)
    plt.text(0.1, 0.3, f'ROC AUC: {roc_auc:.4f}', fontsize=12)
    plt.title('Enhanced Performance Summary')
    
    plt.tight_layout()
    plt.show()
    
    return {
        'genuine_scores': genuine_scores,
        'impostor_scores': impostor_scores,
        'optimal_threshold': optimal_threshold,
        'accuracy': accuracy,
        'eer': eer,
        'roc_auc': roc_auc,
        'd_prime': d_prime
    }
```

```python
# ========================================================
# CELL 6 – RUN COMPLETE PIPELINE
# ========================================================
if __name__ == "__main__":
    print("🚀 STARTING COMPLETE GSNE PIPELINE")
    print("=" * 50)
    
    # Step 1: Process dataset (start with 10 subjects)
    features, labels = process_dataset(max_subjects=108, images_per_eye=7, use_cache=True)
    
    # Step 2: Evaluate matching performance
    if len(features) > 0:
        results = evaluate_matching_performance(features, labels)
        
        # Step 3: Test individual subject matching
        print("\n🔍 INDIVIDUAL SUBJECT MATCHING TEST:")
        print("=" * 40)
        
        unique_subjects = np.unique(labels)
        for subject in unique_subjects[:5]:  # Test first 5 subjects
            subject_indices = np.where(labels == subject)[0]
            if len(subject_indices) >= 2:
                # Calculate within-subject similarities
                subject_features = features[subject_indices]
                within_similarities = []
                for i in range(len(subject_features)):
                    for j in range(i + 1, len(subject_features)):
                        sim = cosine_similarity([subject_features[i]], [subject_features[j]])[0, 0]
                        within_similarities.append(sim)
                
                avg_within_sim = np.mean(within_similarities)
                print(f"👤 Subject {subject}: {len(subject_indices)} images, "
                      f"avg similarity: {avg_within_sim:.4f}")
```
