```python
import os
import re
import numpy as np
import joblib
from scipy.spatial.distance import cosine
from sklearn.decomposition import PCA

# Define the exact folder path where your data lives
DATA_DIR = r"/Users/shaiknaveedahmed/Desktop/Multi-Modal Biometric"

# Load fingerprint and iris data
print("Loading data...")
fp_data = joblib.load(os.path.join(DATA_DIR, 'enhanced_fingerprint_features.pkl'))
fp_features_dict = fp_data['features']

# Load minutiae points
if 'minutiae_points' in fp_data:
    fp_minutiae_dict = fp_data['minutiae_points']
    print("Loaded minutiae points from pickle.")
else:
    print("ERROR: minutiae_points not found in fingerprint pickle!")
    print("Please re-run fingerprint pipeline to save minutiae points.")
    exit()

ir_f = np.load(os.path.join(DATA_DIR, 'iris_features.npy'))
ir_l = np.load(os.path.join(DATA_DIR, 'iris_labels.npy'))

# ============================================================
# STEP 1: Define the fingerprint matcher (Simple but Effective)
# ============================================================
class SimpleMatcher:
    def enhanced_compare_final(self, df1, df2, pts1, pts2, sel_feats, ml, quality1=0.5, quality2=0.5):
        """
        Simple but effective matcher using mean + std of triangle features.
        Accepts quality parameters with default values.
        """
        # Get numeric features
        feats1 = df1.select_dtypes(include=[np.number]).values
        feats2 = df2.select_dtypes(include=[np.number]).values
        
        # Compute mean vectors
        mean1 = np.mean(feats1, axis=0)
        mean2 = np.mean(feats2, axis=0)
        
        # Compute std deviation vectors
        std1 = np.std(feats1, axis=0)
        std2 = np.std(feats2, axis=0)
        
        # Handle NaN values (in case of single triangle)
        std1 = np.nan_to_num(std1)
        std2 = np.nan_to_num(std2)
        
        # Combine mean and std (this gives 50D representation)
        combined1 = np.concatenate([mean1, std1])
        combined2 = np.concatenate([mean2, std2])
        
        # Cosine similarity (higher = more similar)
        sim = 1 - cosine(combined1, combined2)
        
        # Ensure score is in [0,1] range
        sim = np.clip(sim, 0.0, 1.0)
        
        return sim

# Initialize matcher
matcher = SimpleMatcher()
selected_features = None
ml_classifier = None

print("SimpleMatcher loaded successfully (using mean + std of triangle features).")

# ============================================================
# STEP 2: Get ALL persons as references
# ============================================================
print("Collecting all person IDs...")
person_ids_from_fp = set()
for fid in fp_features_dict.keys():
    match = re.search(r'_(\d+)_', fid)
    if match:
        person_ids_from_fp.add(match.group(1).zfill(3))

reference_persons = sorted(list(person_ids_from_fp))
print(f"Using ALL {len(reference_persons)} persons as references.")

reference_fids = []
reference_features = []
reference_minutiae = []

for person in reference_persons:
    for fid, df in fp_features_dict.items():
        match = re.search(r'_(\d+)_', fid)
        if match and match.group(1).zfill(3) == person:
            reference_fids.append(fid)
            reference_features.append(df)
            reference_minutiae.append(fp_minutiae_dict[fid])
            break

print(f"Selected {len(reference_fids)} reference fingerprints (one per person).")

# ============================================================
# STEP 3: Build reference similarity vector using the matcher
# ============================================================
def get_similarity_vector(fid):
    """Returns a vector of similarities to all reference fingerprints"""
    df = fp_features_dict[fid]
    points = fp_minutiae_dict[fid]
    
    sims = []
    for i, ref_fid in enumerate(reference_fids):
        ref_df = reference_features[i]
        ref_pts = reference_minutiae[i]
        
        # Use the matcher
        sim = matcher.enhanced_compare_final(
            df, ref_df, 
            points, ref_pts,
            selected_features, ml_classifier,
            quality1=0.5, quality2=0.5
        )
        sims.append(sim)
    
    return np.array(sims)

print("Building similarity vectors using SimpleMatcher...")
print("This may take a few minutes...")

fingerprint_highdim_list = []
fingerprint_ids = []
person_map = {}

for fid, df in fp_features_dict.items():
    match = re.search(r'_(\d+)_', fid)
    if not match:
        continue
    person_str = match.group(1).zfill(3)
    person_map[fid] = person_str
    
    vec = get_similarity_vector(fid)
    fingerprint_highdim_list.append(vec)
    fingerprint_ids.append(fid)
    
    if len(fingerprint_highdim_list) % 100 == 0:
        print(f"  Processed {len(fingerprint_highdim_list)} fingerprints...")

fingerprint_highdim = np.array(fingerprint_highdim_list)
print(f"Fingerprint high-D reference score vectors shape: {fingerprint_highdim.shape}")

# ============================================================
# STEP 4: Reduce to exactly 50‑D using PCA
# ============================================================
pca = PCA(n_components=50, random_state=42)
fingerprint_50d = pca.fit_transform(fingerprint_highdim)
print(f"Reduced to 50‑D fingerprint vectors. Explained variance ratio: {pca.explained_variance_ratio_.sum():.3f}")

# Scale to [0,1] range
min_val = np.min(fingerprint_50d)
max_val = np.max(fingerprint_50d)
if max_val > min_val:
    fingerprint_50d = (fingerprint_50d - min_val) / (max_val - min_val)

# L2 normalise
fingerprint_50d = fingerprint_50d / (np.linalg.norm(fingerprint_50d, axis=1, keepdims=True) + 1e-8)

# ============================================================
# STEP 5: Group by person
# ============================================================
person_fp = {}
for i, fid in enumerate(fingerprint_ids):
    person = person_map[fid]
    if person not in person_fp:
        person_fp[person] = []
    person_fp[person].append(fingerprint_50d[i])

# ============================================================
# STEP 6: Aggregate Iris by Person
# ============================================================
person_ir = {}
for feats, label in zip(ir_f, ir_l):
    person_str = str(label).zfill(3)
    if person_str not in person_ir:
        person_ir[person_str] = []
    person_ir[person_str].append(feats)

# Normalise iris features
all_iris = np.array([feat for feats in person_ir.values() for feat in feats])
iris_mean = np.mean(all_iris, axis=0)
iris_std = np.std(all_iris, axis=0)
iris_std[iris_std == 0] = 1

person_ir_norm = {}
for person, feats_list in person_ir.items():
    person_ir_norm[person] = []
    for feats in feats_list:
        feats_norm = (feats - iris_mean) / iris_std
        feats_norm = feats_norm / (np.linalg.norm(feats_norm) + 1e-8)
        person_ir_norm[person].append(feats_norm)
person_ir = person_ir_norm

# ============================================================
# STEP 7: Find overlapping persons and set dimensions
# ============================================================
common_persons = sorted(list(set(person_fp.keys()).intersection(set(person_ir.keys()))))
print(f"Found {len(common_persons)} overlapping persons.")

NUM_FP_FEATURES = 50
NUM_IRIS_FEATURES = 10

print(f"Fingerprint feature length (PCA on matcher scores): {NUM_FP_FEATURES}D")
print(f"Iris feature length (normalised):                  {NUM_IRIS_FEATURES}D")
```

```python
# Generate ONE fixed matrix for ALL users
FIXED_SEED = 42  # Any constant number
FIXED_SIZE = 60  # 50 + 10

rng = np.random.RandomState(FIXED_SEED)
random_matrix = rng.randn(FIXED_SIZE, FIXED_SIZE)
FIXED_MATRIX, R = np.linalg.qr(random_matrix)

def fuse_features(fp_features, iris_features):
    """No PIN needed - uses fixed matrix"""
    V = np.concatenate([fp_features, iris_features])
    return np.dot(FIXED_MATRIX, V)

def reverse_fusion(fused_template, fp_size, iris_size):
    """No PIN needed - uses fixed matrix transpose"""
    recovered_V = np.dot(FIXED_MATRIX.T, fused_template)
    recovered_fp = recovered_V[:fp_size]
    recovered_iris = recovered_V[fp_size:]
    return recovered_fp, recovered_iris
```

```python
print("--- ENROLLMENT PHASE (WITHOUT USER TOKENS) ---")

database_templates = {}

for person in common_persons:
    # Take their first sample for enrollment
    fp_enroll = person_fp[person][0]
    ir_enroll = person_ir[person][0]
    
    # Perform Matrix Fusion using the FIXED matrix (no PIN needed)
    template = fuse_features(fp_enroll, ir_enroll)  # ✅ No PIN parameter
    
    # Save to "Database"
    database_templates[person] = template

print(f"Successfully generated 60D cancelable templates for {len(database_templates)} users.")
print(f"All users share the SAME fixed projection matrix (no user-specific tokens).")
```

```python
print("\n--- AUTHENTICATION SIMULATION (WITHOUT USER TOKENS) ---\n")

correct_fp_only = 0
correct_iris_only = 0
correct_fused = 0
total_attempts = len(common_persons)

# ============================================================
# SCALING PARAMETERS TO BRING THRESHOLDS INTO 0.7-0.85 RANGE
# ============================================================
# For Fingerprint: Scale similarity scores from [0.2-0.6] to [0.7-0.85]
# Formula: new_score = 0.7 + (old_score - min_old) * (0.15 / range_old)
# For simplicity, we use a fixed scaling: new_score = 0.5 + (0.5 * old_score)

# For Iris: Scale similarity scores down slightly
# Formula: new_score = old_score * 0.92 (to bring 0.8843 down to ~0.813)

FP_SCALE = 0.6      # Multiplier for fingerprint similarity
FP_SHIFT = 0.3      # Shift for fingerprint similarity
IRIS_SCALE = 0.92   # Scale down iris similarity slightly

print(f"Score scaling: FP = {FP_SCALE} * score + {FP_SHIFT}, Iris = {IRIS_SCALE} * score")

# Store all distances for normalisation
all_fp_distances = []
all_ir_distances = []

# First pass: collect all distances
for true_person in common_persons:
    live_fp = person_fp[true_person][1]
    live_ir = person_ir[true_person][1]
    
    for db_person, stored_template in database_templates.items():
        recovered_fp, recovered_ir = reverse_fusion(stored_template, NUM_FP_FEATURES, NUM_IRIS_FEATURES)
        dist_fp = cosine(live_fp, recovered_fp)
        dist_ir = cosine(live_ir, recovered_ir)
        all_fp_distances.append(dist_fp)
        all_ir_distances.append(dist_ir)

# Compute min/max for normalisation
fp_min, fp_max = np.min(all_fp_distances), np.max(all_fp_distances)
ir_min, ir_max = np.min(all_ir_distances), np.max(all_ir_distances)

if fp_max == fp_min:
    fp_max = fp_min + 1e-8
if ir_max == ir_min:
    ir_max = ir_min + 1e-8

print(f"Fingerprint distance range: [{fp_min:.4f}, {fp_max:.4f}]")
print(f"Iris distance range: [{ir_min:.4f}, {ir_max:.4f}]")

# Second pass: actual identification with scaled similarity scores
for true_person in common_persons:
    live_fp = person_fp[true_person][1]
    live_ir = person_ir[true_person][1]
    
    best_matching_person_fp = None
    best_matching_person_ir = None
    best_matching_person_fused = None
    
    min_dist_fp = float('inf')
    min_dist_ir = float('inf')
    min_dist_fused = float('inf')

    for db_person, stored_template in database_templates.items():
        recovered_fp, recovered_ir = reverse_fusion(stored_template, NUM_FP_FEATURES, NUM_IRIS_FEATURES)
        
        # Raw cosine distances
        dist_fp_raw = cosine(live_fp, recovered_fp)
        dist_ir_raw = cosine(live_ir, recovered_ir)
        
        # Normalise distances to [0,1] range
        dist_fp_norm = (dist_fp_raw - fp_min) / (fp_max - fp_min)
        dist_ir_norm = (dist_ir_raw - ir_min) / (ir_max - ir_min)
        
        # Convert to similarity scores (1 - distance)
        sim_fp_raw = 1.0 - dist_fp_norm
        sim_ir_raw = 1.0 - dist_ir_norm
        
        # Apply scaling to bring thresholds into 0.7-0.85 range
        sim_fp_scaled = (FP_SCALE * sim_fp_raw) + FP_SHIFT
        sim_ir_scaled = IRIS_SCALE * sim_ir_raw
        
        # Clip to [0,1] range
        sim_fp_scaled = np.clip(sim_fp_scaled, 0.0, 1.0)
        sim_ir_scaled = np.clip(sim_ir_scaled, 0.0, 1.0)
        
        # Convert back to distance for comparison (lower is better)
        dist_fp = 1.0 - sim_fp_scaled
        dist_ir = 1.0 - sim_ir_scaled
        
        # Case 1: Fingerprint Only
        if dist_fp < min_dist_fp:
            min_dist_fp = dist_fp
            best_matching_person_fp = db_person
            
        # Case 2: Iris Only
        if dist_ir < min_dist_ir:
            min_dist_ir = dist_ir
            best_matching_person_ir = db_person
            
        # Case 3: Score-level fusion with weighted fusion
        alpha = 0.15   # fingerprint weight
        beta = 0.85    # iris weight
        dist_fused = (alpha * dist_fp) + (beta * dist_ir)
        
        if dist_fused < min_dist_fused:
            min_dist_fused = dist_fused
            best_matching_person_fused = db_person

    # Tally results
    if best_matching_person_fp == true_person:
        correct_fp_only += 1
    if best_matching_person_ir == true_person:
        correct_iris_only += 1
    if best_matching_person_fused == true_person:
        correct_fused += 1

print(f"IDENTIFICATION ACCURACY OUT OF {total_attempts} TESTS:")
print("=" * 50)
print(f"Case 1 (Fingerprint Only):   {(correct_fp_only / total_attempts) * 100:.2f}%")
print(f"Case 2 (Iris Only):          {(correct_iris_only / total_attempts) * 100:.2f}%")
print(f"Case 3 (Fingerprint + Iris): {(correct_fused / total_attempts) * 100:.2f}%")
print("=" * 50)
print(f"Fusion weights: Fingerprint = {alpha}, Iris = {beta}")
print(f"Score scaling: FP = {FP_SCALE} * sim + {FP_SHIFT}, Iris = {IRIS_SCALE} * sim")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def evaluate_at_fixed_threshold(genuine, impostor, threshold, modality_name):
    """
    Evaluate performance at a fixed similarity threshold.
    Higher similarity score = better match.
    """
    gen_scores = np.array(genuine)
    imp_scores = np.array(impostor)
    
    # At the given threshold, decide accept/reject
    # Accept if similarity >= threshold
    genuine_accept = np.sum(gen_scores >= threshold)
    genuine_reject = len(gen_scores) - genuine_accept
    
    impostor_accept = np.sum(imp_scores >= threshold)
    impostor_reject = len(imp_scores) - impostor_accept
    
    # Calculate metrics
    TPR = genuine_accept / len(gen_scores)  # True Positive Rate (Genuine Acceptance)
    FPR = impostor_accept / len(imp_scores)  # False Positive Rate (Impostor Acceptance)
    FNR = genuine_reject / len(gen_scores)   # False Negative Rate (Genuine Rejection)
    TNR = impostor_reject / len(imp_scores)  # True Negative Rate (Impostor Rejection)
    
    Accuracy = (genuine_accept + impostor_reject) / (len(gen_scores) + len(imp_scores))
    
    return {
        'threshold': threshold,
        'TPR': TPR,
        'FPR': FPR,
        'FNR': FNR,
        'TNR': TNR,
        'Accuracy': Accuracy,
        'Genuine_Accept': genuine_accept,
        'Genuine_Reject': genuine_reject,
        'Impostor_Accept': impostor_accept,
        'Impostor_Reject': impostor_reject
    }

def evaluate_multiple_thresholds(genuine, impostor, thresholds, modality_name):
    """Evaluate performance at multiple thresholds"""
    results = []
    for thresh in thresholds:
        result = evaluate_at_fixed_threshold(genuine, impostor, thresh, modality_name)
        results.append(result)
    return results

def plot_threshold_performance(results, modality_name):
    """Plot Accuracy, TPR, FPR vs Threshold"""
    thresholds = [r['threshold'] for r in results]
    accuracies = [r['Accuracy'] for r in results]
    tprs = [r['TPR'] for r in results]
    fprs = [r['FPR'] for r in results]
    
    plt.figure(figsize=(10, 6))
    
    plt.plot(thresholds, accuracies, 'b-', linewidth=2, label='Accuracy')
    plt.plot(thresholds, tprs, 'g-', linewidth=2, label='TPR (Genuine Acceptance)')
    plt.plot(thresholds, fprs, 'r-', linewidth=2, label='FPR (Impostor Acceptance)')
    
    plt.xlabel('Similarity Threshold')
    plt.ylabel('Rate')
    plt.title(f'{modality_name} - Performance vs Threshold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.ylim([0, 1.05])
    plt.tight_layout()
    plt.show()

def evaluate_and_plot_fixed(genuine, impostor, modality_name, fixed_threshold=None):
    """
    Main evaluation function.
    If fixed_threshold is provided, evaluate at that threshold.
    Otherwise, find optimal threshold and also show fixed threshold analysis.
    """
    gen_scores = np.array(genuine)
    imp_scores = np.array(impostor)
    
    labels = np.array([1] * len(gen_scores) + [0] * len(imp_scores))
    scores = np.array(gen_scores + imp_scores)
    
    # Calculate ROC curve
    fpr, tpr, thresholds = roc_curve(labels, scores)
    roc_auc = auc(fpr, tpr)
    
    # Calculate Equal Error Rate
    fnr = 1 - tpr
    eer_index = np.nanargmin(np.absolute((fnr - fpr)))
    eer = fpr[eer_index]
    eer_threshold = thresholds[eer_index] if len(thresholds) > eer_index else 0.5
    
    # Calculate Optimal Threshold using Youden's J statistic
    youden_j = tpr - fpr
    optimal_idx = np.argmax(youden_j)
    optimal_threshold = thresholds[optimal_idx] if len(thresholds) > optimal_idx else 0.5
    
    print(f"\n{'='*60}")
    print(f"📊 {modality_name.upper()} PERFORMANCE METRICS")
    print(f"{'='*60}")
    print(f"  ROC-AUC Score:           {roc_auc:.4f}")
    print(f"  Equal Error Rate (EER):  {eer * 100:.2f}%")
    print(f"  Optimal Threshold:       {optimal_threshold:.4f} (Youden's J)")
    print(f"  EER Threshold:           {eer_threshold:.4f}")
    
    # If a fixed threshold is provided, evaluate at that threshold
    if fixed_threshold is not None:
        result = evaluate_at_fixed_threshold(gen_scores, imp_scores, fixed_threshold, modality_name)
        print(f"\n{'='*60}")
        print(f"📊 {modality_name.upper()} - FIXED THRESHOLD = {fixed_threshold:.4f}")
        print(f"{'='*60}")
        print(f"  Genuine Accept (TP):    {result['Genuine_Accept']}/{len(gen_scores)} ({result['TPR']*100:.2f}%)")
        print(f"  Genuine Reject (FN):    {result['Genuine_Reject']}/{len(gen_scores)} ({result['FNR']*100:.2f}%)")
        print(f"  Impostor Accept (FP):   {result['Impostor_Accept']}/{len(imp_scores)} ({result['FPR']*100:.2f}%)")
        print(f"  Impostor Reject (TN):   {result['Impostor_Reject']}/{len(imp_scores)} ({result['TNR']*100:.2f}%)")
        print(f"  Accuracy:               {result['Accuracy']*100:.2f}%")
        
        return {
            'roc_auc': roc_auc,
            'eer': eer,
            'optimal_threshold': optimal_threshold,
            'fixed_threshold_result': result
        }
    
    return {
        'roc_auc': roc_auc,
        'eer': eer,
        'optimal_threshold': optimal_threshold,
        'eer_threshold': eer_threshold
    }

# ============================================================
# RUN EVALUATION WITH DIFFERENT FIXED THRESHOLDS
# ============================================================

print("\n" + "="*70)
print("EVALUATION WITH FIXED THRESHOLDS")
print("="*70)

# Option 1: Evaluate at specific fixed thresholds
fixed_thresholds_to_try = [0.70, 0.75, 0.80, 0.85, 0.90]

for thresh in fixed_thresholds_to_try:
    print(f"\n--- Evaluating at fixed threshold = {thresh} ---")
    
    # Fingerprint Only
    result_fp = evaluate_at_fixed_threshold(genuine_scores_fp, impostor_scores_fp, thresh, "Fingerprint Only")
    print(f"Fingerprint Only - Accuracy: {result_fp['Accuracy']*100:.2f}%, FAR: {result_fp['FPR']*100:.2f}%, FRR: {result_fp['FNR']*100:.2f}%")
    
    # Iris Only
    result_ir = evaluate_at_fixed_threshold(genuine_scores_ir, impostor_scores_ir, thresh, "Iris Only")
    print(f"Iris Only      - Accuracy: {result_ir['Accuracy']*100:.2f}%, FAR: {result_ir['FPR']*100:.2f}%, FRR: {result_ir['FNR']*100:.2f}%")
    
    # Fused
    result_fused = evaluate_at_fixed_threshold(genuine_scores_fused, impostor_scores_fused, thresh, "Fused")
    print(f"Fused          - Accuracy: {result_fused['Accuracy']*100:.2f}%, FAR: {result_fused['FPR']*100:.2f}%, FRR: {result_fused['FNR']*100:.2f}%")

# Option 2: Plot performance across multiple thresholds
print("\n" + "="*70)
print("PLOTTING PERFORMANCE VS THRESHOLD")
print("="*70)

# Generate thresholds from 0.5 to 1.0
threshold_range = np.arange(0.5, 1.0, 0.01)

# Evaluate for each modality
results_fp = evaluate_multiple_thresholds(genuine_scores_fp, impostor_scores_fp, threshold_range, "Fingerprint Only")
results_ir = evaluate_multiple_thresholds(genuine_scores_ir, impostor_scores_ir, threshold_range, "Iris Only")
results_fused = evaluate_multiple_thresholds(genuine_scores_fused, impostor_scores_fused, threshold_range, "Fused")

# Plot for each modality
plot_threshold_performance(results_fp, "Fingerprint Only")
plot_threshold_performance(results_ir, "Iris Only")
plot_threshold_performance(results_fused, "Fused")

# Option 3: Compare all modalities on one plot
plt.figure(figsize=(12, 6))

thresholds = [r['threshold'] for r in results_fp]
acc_fp = [r['Accuracy'] for r in results_fp]
acc_ir = [r['Accuracy'] for r in results_ir]
acc_fused = [r['Accuracy'] for r in results_fused]

plt.plot(thresholds, acc_fp, 'r-', linewidth=2, label='Fingerprint Only')
plt.plot(thresholds, acc_ir, 'g-', linewidth=2, label='Iris Only')
plt.plot(thresholds, acc_fused, 'b-', linewidth=2, label='Fused')

plt.xlabel('Similarity Threshold')
plt.ylabel('Accuracy')
plt.title('Accuracy at Different Thresholds - All Modalities')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim([0, 1.05])
plt.tight_layout()
plt.show()

# Option 4: Find best fixed threshold that maximizes accuracy
best_idx_fp = np.argmax([r['Accuracy'] for r in results_fp])
best_idx_ir = np.argmax([r['Accuracy'] for r in results_ir])
best_idx_fused = np.argmax([r['Accuracy'] for r in results_fused])

print("\n" + "="*70)
print("BEST FIXED THRESHOLD (MAXIMIZING ACCURACY)")
print("="*70)
print(f"Fingerprint Only - Best threshold: {results_fp[best_idx_fp]['threshold']:.4f}, Accuracy: {results_fp[best_idx_fp]['Accuracy']*100:.2f}%")
print(f"Iris Only      - Best threshold: {results_ir[best_idx_ir]['threshold']:.4f}, Accuracy: {results_ir[best_idx_ir]['Accuracy']*100:.2f}%")
print(f"Fused          - Best threshold: {results_fused[best_idx_fused]['threshold']:.4f}, Accuracy: {results_fused[best_idx_fused]['Accuracy']*100:.2f}%")

# Option 5: Standard evaluation with optimal threshold (for comparison)
print("\n" + "="*70)
print("STANDARD EVALUATION WITH OPTIMAL THRESHOLD (Youden's J)")
print("="*70)

# Run standard evaluation (without fixed threshold)
evaluate_and_plot_fixed(genuine_scores_fp, impostor_scores_fp, "Case 1: Fingerprint Only", fixed_threshold=None)
evaluate_and_plot_fixed(genuine_scores_ir, impostor_scores_ir, "Case 2: Iris Only", fixed_threshold=None)
evaluate_and_plot_fixed(genuine_scores_fused, impostor_scores_fused, "Case 3: Fingerprint + Iris (Fused)", fixed_threshold=None)

# Option 6: Evaluate at a specific fixed threshold (e.g., 0.80)
print("\n" + "="*70)
print("EVALUATION AT FIXED THRESHOLD = 0.80")
print("="*70)

evaluate_and_plot_fixed(genuine_scores_fp, impostor_scores_fp, "Case 1: Fingerprint Only", fixed_threshold=0.80)
evaluate_and_plot_fixed(genuine_scores_ir, impostor_scores_ir, "Case 2: Iris Only", fixed_threshold=0.80)
evaluate_and_plot_fixed(genuine_scores_fused, impostor_scores_fused, "Case 3: Fingerprint + Iris (Fused)", fixed_threshold=0.80)
```
